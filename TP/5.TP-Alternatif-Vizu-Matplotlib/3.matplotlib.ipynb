### Subplot

Une figure matplotlib peut contenir plusieurs sous-plans. 
Ces sous-plots sont organisés dans une grille. Pour créer un sous-plot,
il suffit d'appeler la fonction subplot, et de spécifier le nombre de lignes 
et de colonnes de la figure, ainsi que l'indice du sous-plot sur lequel vous
souhaitez dessiner (en commençant par 1, puis de gauche à droite, et de haut en bas)   

```python
x = np.linspace(-1.4,1.4,30)

plt.subplot(2,2,1)
plt.plot(x,x)
plt.title('lineaire')

plt.subplot(2,2,2)
plt.plot(x,x**2)
plt.title('carree')

plt.subplot(2,2,3)
plt.plot(x,x**3)
plt.title('cube')

plt.subplot(2,2,4)
plt.plot(x,x**4)
plt.title('puissance 4')

plt.show()
```


Notez que $\textbf{subplot(223)}$ est un raccourci pour $\textbf{subplot(2, 2, 3)}$.


On peut aussi créer des sous-plots qui s'étendent sur plusieurs cellules de la grille, comme suit :

```python
plt.subplot(2,2,1)
plt.plot(x,x)
plt.title('lineaire')

plt.subplot(2,2,2)
plt.plot(x,x**2)
plt.title('carree')

plt.subplot(2,1,2)
plt.plot(x,x**3)
plt.title('cube')
plt.show()
```

```python
fig, ax = plt.subplots(2,1)

y = x**2
ax[0].plot(x,y)
ax[1].plot(x,y)

plt.show()
```

```python
# Reprise de l'exemple du poly

def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)

t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0, 0.02)

fig, (ax1, ax2) = plt.subplots(2, 1)
ax1.plot(t1, f(t1), 'bo', t2, f(t2),'k')
ax2.plot(t2, np.cos(2*np.pi*t2), 'r--', linewidth=2.0)
```

```python
# un autre exemple

x = np.linspace(-1.4, 1.4, 30)
line1, line2, line3 = plt.plot(x, x, 'g--', x, x**2, 'r:', x, x**3, 'b^')

line1.set_linewidth(3.0)
line2.set_linewidth(3.0)
line1.set_dash_capstyle("round")
line3.set_alpha(0.2)
plt.grid()
plt.show()



```

$\textbf{exercice}$  : representer sur un même graphique plusieurs fonctions qui sont le resultat de
 plusieurs experiences

```python
# on commence par créé des données 

nbre_total_exp = 4
dataset = {f"experience{i}":np.random.randn(100,3) for i in range(nbre_total_exp)}
#dataset
```

On fait ensuite une petite fonction qui nous permettra de les afficher. Cette fonction peut etre tres utile pour le reste


```python
def graphique(data):
    n = len(data)
    plt.figure(figsize=(12,8))
    for k , i in zip(data.keys(), range(1,n+1)):
        plt.subplot(n,1,i)
        plt.plot(data[k])
        plt.title(k)
    plt.show()
```

```python
#on affiche nos données

graphique(dataset)

# on peut modifier le nombre total d'experience dans nbre_total_exp et le code qui 
# affiche continuera de fonctionner


```

## Dessin de texte

Vous pouvez ajouter du texte à n'importe quel endroit du graphique. Il suffit de spécifier les coordonnées horizontales et verticales et le texte, et éventuellement quelques attributs supplémentaires. Tout texte dans matplotlib peut contenir des expressions d'équation TeX, (voir la documentation pour plus de détails).


```python
# un exemple

x = np.linspace(-1.5, 1.5, 30)
px = 0.8
py = px**2

plt.plot(x, x**2, "b-", px, py, "ro")

plt.text(0, 1.5, "function carree\n$y = x^2$", fontsize=20, color='blue', horizontalalignment="center")
plt.text(px - 0.08, py, " un point", ha="right", weight="heavy")
plt.text(px, py, "x = %0.2f\ny = %0.2f"%(px, py), rotation=50, color='gray')

plt.grid()
plt.show()
```

- Remarque : ha est un alias de horizontalalignment.

Pour d'autres propriétés de texte, consultez la documentation.

Il est assez fréquent d'annoter des éléments d'un graphique, comme le beau point ci-dessus. 
La fonction d'annotation facilite cette tâche :
il suffit d'indiquer l'emplacement du point d'intérêt et la position du texte, 
plus éventuellement quelques attributs supplémentaires pour le texte et la flèche.




```python
plt.plot(x, x**2, px, py, "ro")
plt.annotate("un point", xy=(px, py), xytext=(px-1.3,py+0.5),
                           color="green", weight="heavy", fontsize=14,
                           arrowprops={"facecolor": "lightgreen"})

plt.grid()
plt.show()
```

Vous pouvez également ajouter une boîte de délimitation autour de votre texte en utilisant l'attribut bbox :


```python

plt.plot(x, x**2, px, py, "ro")

bbox_props = dict(boxstyle="rarrow,pad=0.3", ec="b", lw=2, fc="lightblue")
plt.text(px-0.2, py, "Un point", bbox=bbox_props, ha="right")

bbox_props = dict(boxstyle="round4,pad=1,rounding_size=0.2", ec="black", fc="#EEEEFF", lw=5)
plt.text(0, 1.5, "fonction carree\n$y = x^2$", fontsize=20, color='black', ha="center", bbox=bbox_props)

plt.grid()

plt.show()
```

## Le nuage de points: scatter

Pour dessiner un nuage de points, il suffit de fournir les coordonnées x et y des points.

```python
x,y = np.random.rand(2,100)
plt.scatter(x,y)

plt.show()
```

Vous pouvez également indiquer l'échelle de chaque point.


```python
x,y,scale = np.random.rand(3,100)
scale = 500*scale**5
plt.scatter(x,y,s=scale)
plt.show()
```

Et comme d'habitude, vous pouvez définir un certain nombre d'autres attributs, tels que les couleurs de remplissage et de bordure et le niveau alpha.

```python
for color in ['red','green','blue']:
    n = 100
    x,y = np.random.rand(2,n)
    scale=500*np.random.rand(n)**5
    plt.scatter(x,y,s =scale,c=color, alpha=0.3)
    
plt.grid()
plt.show()
```

## Histogramme

Ils sont tres importants lorsqu'on manipule les données. Ils nous permmetent 
de voir la distribution de nos donnees . Si la distribution est normale ou non, symétrique ou asymétrique ou bien encore de visualiser la moyennne. On peut voir comment les fleurs sont distribuées

```python
# un exemple avec la fleur d'iris avec modification du nombre de bins

plt.hist(X[:,0], bins=30)
plt.hist(X[:,1], bins=30)
plt.grid()
plt.show()
```

```python
#un autre exemple
data = [1, 1.1, 1.8,2,2.1,3.2,3,3,3,3,3]

plt.subplot(211)
plt.hist(data, bins = 10)

plt.subplot(212)
plt.hist(data, bins = [1, 1.5,2, 2.5,3])

plt.show()
```

On peut dessiner des histogrammes sur le même graphique (reprise fleur d'iris avec les deux premières données)




```python
data1 = np.random.randn(400)
data2 = np.random.randn(500) + 3
data3 = np.random.randn(450) + 6
data4 = np.random.randn(200) + 9
data5 = np.random.randn(100) + 10


plt.hist(data1, bins=5, color='g', alpha=0.75, label='bar hist') # par defaut
plt.hist(data2, color='b', alpha=0.5, histtype= 'stepfilled',label='stepfilled hist')
plt.hist(data3, color='r', alpha=0.65, histtype= 'step',label='step hist')
plt.hist((data4, data5), color=('r','m'), alpha=0.55, histtype= 'barstacked',
         label=('barstacked 1', 'barstacked 2'))



plt.legend()
plt.show()
```

Il y a aussi différents types d'histogrammmes et un certains nombres de paramètres qu'on peut modifier

On peut créer notre propre image

Comme nous n'avons pas fourni de niveaux RGB, la fonction imshow fait automatiquement correspondre les valeurs à un dégradé de couleurs. Par défaut, le gradient de couleur va du bleu (pour les valeurs faibles) au rouge (pour les valeurs élevées), mais vous pouvez sélectionner une autre carte de couleurs. Par exemple :


```python
img = np.empty((20,30,3))
img[:,:10] = [0,0,0.6]
img[:,10:20] = [1,1,1]
img[:,20:] = [0.6,0,0]

plt.imshow(img)
plt.show()
```

Vous pouvez souhaiter masquer les axes lorsque vous affichez une image :

```python
plt.imshow(img)
plt.axis('off')
plt.show()
```

On peut generer notre propre image

Puisque le tableau img est tout simplement assez petit (20x30), lorsque la fonction imshow l'affiche, elle agrandit l'image à la taille de la figure. Par défaut, elle utilise l'interpolation bilinéaire pour remplir les pixels ajoutés. C'est pourquoi les bords semblent flous. Vous pouvez sélectionner un autre algorithme d'interpolation, par exemple en copiant la couleur du pixel le plus proche :


```python
img = np.empty((20,30,3))
img[:,:10] = [0,0,0.6]
img[:,10:20] = [1,1,1]
img[:,20:] = [0.6,0,0]

plt.imshow(img, interpolation="nearest")
plt.axis('off')
plt.show()
```

## Figure en 3D

Pour tracer des graphiques en 3D vous devez importer Axes3D, qui enregistre la projection "3d". Ensuite, créez un sous-plot en fixant la projection à "3d". Cela renvoie un objet Axes3DSubplot, que vous pouvez utiliser pour appeler plot_surface, en donnant des coordonnées x, y et z, ainsi que des attributs optionnels.

```python
# exemple avec la fleur d'iris

from mpl_toolkits.mplot3d import Axes3D
from sklearn.datasets import load_iris


iris = load_iris()
X = iris.data
y = iris.target


ax = plt.axes(projection = '3d')
ax.scatter(X[:,0],X[:,1],X[:,2])
plt.show()
```

```python
# un autre exemple
x = np.linspace(-5,5,50)
y = np.linspace(-5,5,50)
X,Y= np.meshgrid(x,y)


R = np.sqrt(X**2 +Y**2)
Z = np.sin(R)


figure = plt.figure(1, figsize = (12,4))
subplot3d = plt.subplot(111, projection='3d')
surface = subplot3d.plot_surface(X,Y,Z, cmap= plt.cm.coolwarm)
plt.show()
```

Une autre façon d'afficher ces mêmes données est d'utiliser un tracé de contour.

```python
# un autre exemple pour finir :  une surface
plt.contour(X,Y,Z)
plt.colorbar()
plt.show()
```

```python
ax = plt.axes(projection = '3d')
ax.plot_surface(X,Y,Z, cmap="plasma")
```

Vue de dessus d'un 3D :  contour plot



```python

f = lambda x, y : np.sin(x) + np.cos(x+y)*np.cos(x)


X = np.linspace(0,5,200)
Y = np.linspace(0,5,200)


XX,YY = np.meshgrid(X,Y)


Z = f(XX,YY)
```

```python
plt.contour(XX,YY,Z,20,colors='black')

plt.show()


plt.contourf(XX,YY,Z,20,cmap='RdGy')
plt.colorbar()

plt.show()
```
